```python
import requests
import urllib.request
import re
import time
from bs4 import BeautifulSoup as bs

c = []
e2 = [] # d
e3 = [] # e

for k in range(len(b)):#len(b)
    try:
        msoup = bs(requests.get(b[k]).content, "html.parser")
        f = msoup.find("a",class_='next_end')['onclick'] # 텍스트에 페이지 수가 담겨 있는 태그 추출
        f2 = f.split('(')
        f3 = f2[1].split(',')
        maxPrice = f3[0]
    except TypeError:
        print('분석을 위한 후기의 개수가 부족한 제품입니다.')
    for page_number in range(int(maxPrice)+1):
        c.append(b[k].replace("detail.nhn?cat_id=50000437&nv_mid=","review_list.nhn?nvMid=") + "&page=" + str(page_number) +"&reviewSort=accuracy&reviewType=all&ligh=true")
        d1 = requests.post(c[page_number-1])
        d2 = d1.text
        d3 = bs(d2, 'html.parser')
        d4 = d3.find_all('div','atc_area') # 후기 부분 필터링
        for d5 in d4:
            if int(d5.find('div','avg_area').find('span','curr_avg').getText()) == 4 or int(d5.find('div','avg_area').find('span','curr_avg').getText()) == 5:
                e2.append(d5.find('div','atc')) # 평점 4점 이상의 후기 수집
            elif int(d5.find('div','avg_area').find('span','curr_avg').getText()) == 1 or int(d5.find('div','avg_area').find('span','curr_avg').getText()) == 2:
                e3.append(d5.find('div','atc')) # 평점 2점 이하의 후기 수집
            
pk = []
for i in range(len(e2)):
    pk.append(re.sub('<.+?>', '', str(e2[i]), 0).strip())

nk = []
for i in range(len(e3)):
    nk.append(re.sub('<.+?>', '', str(e3[i]), 0).strip()) 
            
print(nk)
```

```python
review_list1=[]
review_list2=[]

for i in pk:
        review_list1.append(i.strip().replace('\n','').replace('\t','').replace('\r',''))

for i in nk:
        review_list2.append(i.strip().replace('\n','').replace('\t','').replace('\r',''))

# txt 추출
# file1 = open('Positive Data.txt','w',encoding='utf-8')
# file2 = open('Negative Data.txt','w',encoding='utf-8')

# for i in review_list1:
#     file1.write(i+'\n')

# for i in review_list2:
#     file2.write(i+'\n')

# file1.close()
# file2.close()

# json 추출
js1 = json.dumps(review_list1, indent=4)
f1 = open('Positive Data.json', "w")
f1.write(js1)
f1.close()

js2 = json.dumps(review_list2, indent=4)
f2 = open('Negative Data.json', "w")
f2.write(js2)
f2.close()
```
